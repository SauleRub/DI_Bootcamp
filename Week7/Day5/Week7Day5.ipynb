# Exercise 1: Structured vs. Unstructured Data

# 1. A company’s financial reports stored in an Excel file → Structured
# 2. Photographs uploaded to a social media platform → Unstructured
# 3. A collection of news articles on a website → Unstructured
# 4. Inventory data in a relational database → Structured
# 5. Recorded interviews from a market research study → Unstructured


# Exercise 2: Transforming Unstructured Data

# 1. Blog posts → Use NLP to extract entities (location, date) and sentiment analysis to structure key points.
# 2. Audio recordings → Use speech-to-text tools (e.g., Google Speech API) to transcribe, then analyze.
# 3. Handwritten notes → Use OCR (Optical Character Recognition) like Tesseract to convert to text, then organize.
# 4. Cooking video → Use video-to-text narration tools and frame tagging to identify steps and ingredients.
