## Exercise 5: Basic Data Analysis with the Iris Dataset

```python
print("Mean:", iris_df["sepal_length"].mean())
print("Median:", iris_df["sepal_length"].median())
print("Mode:", iris_df["sepal_length"].mode()[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(iris_df["sepal_length"], bins=10, kde=True, color="skyblue")
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

###  Exercise 5: Analysis of Sepal Length

- The mean and median are close, indicating a roughly symmetrical distribution.
- The mode shows the most frequent sepal length value.
- The histogram reveals a bell-shaped curve, suggesting a normal distribution.
- No extreme outliers are visible in this column.

```python
import matplotlib.pyplot as plt
import seaborn as sns

grouped = sleep_df.groupby(["Age Group", "Sex"])["Avg hrs per day sleeping"].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=grouped, x="Age Group", y="Avg hrs per day sleeping", hue="Sex", palette="Set2")

plt.title("Average Sleep Hours by Age Group and Sex")
plt.xlabel("Age Group")
plt.ylabel("Average Hours per Day")
plt.xticks(rotation=45)
plt.legend(title="Sex")
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```


The bar chart reveals several interesting patterns:

1. **Age and Sleep Duration:**
   - Younger individuals (15â€“24) average the most sleep (around 9 hours per night).
   - Sleep duration declines with age, particularly after 25, reaching a stable range by middle age.

2. **Sex-Based Differences:**
   - Women tend to sleep slightly more than men across all age groups.
   - The "Both" category sits consistently between the two.

3. **Overall Insight:**
   - The chart highlights both age and gender as important factors in sleep behavior.
   - The changes are subtle but consistent, showing predictable human sleep trends across life stages.
