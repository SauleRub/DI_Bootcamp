```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = "flower_data/Flower Color Images Dataset/flowers/flowers/19_010.png"

original_image = Image.open(image_path)
plt.imshow(original_image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
from PIL import ImageOps

horizontally_flipped = ImageOps.mirror(original_image)

vertically_flipped = ImageOps.flip(original_image)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].imshow(horizontally_flipped)
axs[0].axis("off")
axs[0].set_title("Flipped Horizontally")

axs[1].imshow(vertically_flipped)
axs[1].axis("off")
axs[1].set_title("Flipped Vertically")

plt.tight_layout()
plt.show()
```

```python
width, height = original_image.size
zoomed_image = original_image.resize((int(width * 1.2), int(height * 1.2)))

plt.imshow(zoomed_image)
plt.axis("off")
plt.title("Zoomed In (120%)")
plt.show()
```
