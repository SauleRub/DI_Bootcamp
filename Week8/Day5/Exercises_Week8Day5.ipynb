##  Exercise 1: Understanding Data Visualization

### Why is Data Visualization Important?
Data visualization is crucial in data analysis because it helps:
- Reveal patterns, trends, and outliers that might be hidden in raw data.
- Communicate complex insights in an intuitive and clear format.
- Allow stakeholders to make data-driven decisions with greater understanding.
- Speed up the process of analysis by allowing visual comparisons and exploration.

### Purpose of a Line Graph
A line graph is used to display information that changes continuously over time.
It is especially useful for:
- Showing trends over intervals (e.g., days, months, years)
- Comparing multiple data sets over the same time frame
- Highlighting rates of increase or decrease

## Exercise 2: Line Plot for Temperature Variation

```python
import matplotlib.pyplot as plt

# Data
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperatures = [20, 25, 26, 24, 22, 18, 16]

# Plot
plt.plot(days, temperatures, marker='o', color='orange')
plt.title("Weekly Temperature Variation")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import os
print(os.listdir("sales_data"))
```

```python
import pandas as pd

df = pd.read_csv("sales_data/sales_data.csv")
df.head()
```

```python
total_quantity = df["quantity"].sum()
print("Total quantity sold:", total_quantity)
```

```python
revenue_by_category = df.groupby("category")["revenue"].sum()
highest_category = revenue_by_category.idxmax()
highest_revenue = revenue_by_category.max()

print("Top category:", highest_category)
print("Revenue generated:", highest_revenue)
```

```python
avg_revenue = df["revenue"].mean()
print("Average revenue per sale:", avg_revenue)
```

```python
df["date"] = pd.to_datetime(df["date"])
df["quarter"] = df["date"].dt.quarter
```

```python
import matplotlib.pyplot as plt

revenue_by_quarter = df.groupby("quarter")["revenue"].sum()

plt.figure(figsize=(8, 5))
plt.bar(revenue_by_quarter.index, revenue_by_quarter.values, color="salmon")

plt.title("Total Revenue by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Revenue ($)")
plt.xticks([1, 2, 3, 4], labels=["Q1", "Q2", "Q3", "Q4"])
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

## Exercise 5: Data Visualization Using Matplotlib

```python
import matplotlib.pyplot as plt

x = list(range(-10, 11))
y = [i**2 for i in x]

plt.plot(x, y, marker='o')
plt.title("Line Plot of y = xÂ²")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
products = ["A", "B", "C", "D"]
sales = [15, 30, 45, 20]

plt.bar(products, sales, color="cornflowerblue")
plt.title("Weekly Sales of Products")
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

```python
labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [40, 30, 20, 10]
colors = ["red", "gold", "pink", "brown"]

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=140)
plt.title("Favorite Fruits")
plt.legend(title="Fruits")
plt.axis("equal")
plt.show()
```
