Exercise 2

```python
import numpy as np
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

# Mean
mean = np.mean(data)

# Median
median = np.median(data)

# Variance
variance = np.var(data, ddof=1)  # sample variance

# Standard Deviation
std_dev = np.std(data, ddof=1)  # sample std

print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

Exercise 3

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generating values from a normal distribution
mean = 50
std_dev = 10
x = np.linspace(0, 100, 1000)
y = norm.pdf(x, mean, std_dev)

# Plot
plt.plot(x, y)
plt.title('Normal Distribution (μ = 50, σ = 10)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

Exercise 4

```python
import numpy as np
from scipy.stats import ttest_ind

# Generating two sets of data (normal distribution)
np.random.seed(42)  # for reproducibility
group1 = np.random.normal(loc=60, scale=8, size=30)
group2 = np.random.normal(loc=65, scale=10, size=30)

# Independent t-test
t_stat, p_value = ttest_ind(group1, group2)

# Output the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("There is a significant difference between the two groups.")
else:
    print("There is no significant difference between the two groups.")
```

```python
import pandas as pd
from scipy import stats
import numpy as np

df = pd.read_csv('Iris.csv')

print("Preview of dataset:")
print(df.head())

sepal_length = df['SepalLengthCm']

from scipy import stats

mean_val = stats.tmean(sepal_length)
median_val = np.median(sepal_length)
variance_val = stats.tvar(sepal_length)
std_dev_val = stats.tstd(sepal_length)

print("\nDescriptive Statistics for 'sepal_length':")
print("Mean:", mean_val)
print("Median:", median_val)
print("Variance:", variance_val)
print("Standard Deviation:", std_dev_val)
```

```python
import numpy as np
from scipy.stats import kurtosis, skew

# Generate random data from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate kurtosis and skewness
kurt = kurtosis(data)
skewness = skew(data)

# Print results
print("Kurtosis:", kurt)
print("Skewness:", skewness)
```
