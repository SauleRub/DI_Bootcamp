```python
import numpy as np
from scipy import stats

# Mean, median, mode
means = df.mean()
medians = df.median()
modes = df.mode().iloc[0]

print("Means:\n", means)
print("\nMedians:\n", medians)
print("\nModes:\n", modes)
```

```python
ranges = df.max() - df.min()
variances = df.var()
std_devs = df.std()

print("\nRanges:\n", ranges)
print("\nVariances:\n", variances)
print("\nStandard Deviations:\n", std_devs)
```

```python
skewness = df.skew()
kurtosis = df.kurtosis()

print("\nSkewness:\n", skewness)
print("\nKurtosis:\n", kurtosis)
```

```python
group0 = df[df['price_range'] == 0]['battery_power']
group3 = df[df['price_range'] == 3]['battery_power']

t_stat, p_value = stats.ttest_ind(group0, group3)
print(f"\nT-statistic: {t_stat}, P-value: {p_value}")
```

```python
correlations = df.corr()['price_range'].sort_values(ascending=False)
print("\nCorrelation with target (price_range):\n", correlations)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))

sns.histplot(df['battery_power'], bins=40, stat='density', kde=False, color='skyblue')

sns.kdeplot(df['battery_power'], color='blue', linewidth=2)

plt.title('Distribution of Battery Power')
plt.xlabel('Battery Power (mAh)')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Scatter plot
sns.scatterplot(data=df, x='battery_power', y='price_range', hue='price_range', palette='viridis')

# Styling
plt.title('Battery Power vs. Price Range')
plt.xlabel('Battery Power (mAh)')
plt.ylabel('Price Range')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='price_range', y='battery_power', data=df, palette='pastel')

plt.title('Battery Power Distribution by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Battery Power (mAh)')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df.corr()

plt.figure(figsize=(12, 8))

sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Matrix Heatmap")

plt.tight_layout()
plt.show()
```

```python
print("🔍 Summary of Key Statistical Metrics:\n")

print(f"Battery Power — Mean: {df['battery_power'].mean():.2f}, Std: {df['battery_power'].std():.2f}")
print(f"RAM — Mean: {df['ram'].mean():.2f}, Std: {df['ram'].std():.2f}")
print(f"Talk Time — Mean: {df['talk_time'].mean():.2f}, Std: {df['talk_time'].std():.2f}")
```

```python
# Correlation with target
correlation = df.corr(numeric_only=True)['price_range'].sort_values(ascending=False)

print("🔥 Features most correlated with price_range:\n")
print(correlation.head(5))
print("\n🧊 Features least correlated with price_range:\n")
print(correlation.tail(5))
```
