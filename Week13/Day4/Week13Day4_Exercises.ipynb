🌟 Exercise 2: Scraping robots.txt from Wikipedia in Google Colab.

```python
import requests

# Define the URL of robots.txt
url = "https://en.wikipedia.org/robots.txt"

# Send a GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("robots.txt content:\n")
    print(response.text)
else:
    print("Failed to fetch robots.txt. Status code:", response.status_code)
```

🌟 Exercise 3: Extracting Headers from Wikipedia’s Main Page.

```python
import requests
from bs4 import BeautifulSoup

# Fetch Wikipedia's main page
url = "https://en.wikipedia.org/wiki/Main_Page"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Extract all header tags from h1 to h6
header_tags = []
for i in range(1, 7):
    headers = soup.find_all(f"h{i}")
    header_tags.extend(headers)

# Display headers
print("Header tags found on Wikipedia's main page:\n")
for tag in header_tags:
    print(f"{tag.name}: {tag.get_text(strip=True)}")
```

🌟 Exercise 4: Checking for Page Title.

```python
import requests
from bs4 import BeautifulSoup

# URL to check
url = "https://en.wikipedia.org/wiki/Main_Page"

# Fetch and parse the page
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Check and print title
title_tag = soup.title

if title_tag:
    print("✅ Page contains a <title> tag.")
    print("Title content:", title_tag.string)
else:
    print("❌ No <title> tag found on the page.")
```


🌟 Exercise 5 : Analyzing US-CERT Security Alerts

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime

# URL of the US-CERT RSS feed
url = "https://www.cisa.gov/news-events/alerts/rss.xml"

# Fetch the RSS feed content
response = requests.get(url)
soup = BeautifulSoup(response.content, features="xml")

# Get current year
current_year = datetime.now().year

# Find all items and count how many match the current year
alerts = soup.find_all('item')
count_this_year = 0

for alert in alerts:
    pub_date = alert.find('pubDate')
    if pub_date and str(current_year) in pub_date.text:
        count_this_year += 1

print(f"📢 Number of US-CERT alerts in {current_year}: {count_this_year}")
```

🌟 Exercise 6: Scraping IMDb (rotten tomatoes) – Top 10 Random Movies

```python
import requests
from bs4 import BeautifulSoup

url = "https://editorial.rottentomatoes.com/guide/best-movies-of-all-time/"

# Fetch the page
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extract movie blocks
movie_blocks = soup.select("div.article_movie_title")

print("🍅 Rotten Tomatoes Top 10 Movies:\n")
for i, block in enumerate(movie_blocks[:10]):
    title = block.h2.a.text.strip()
    print(f"{i+1}. 🎬 {title}")
```
